```python
bool_ChargeOff = df_y.ChargeOff ==1
bool_NotChargeOff = df_y.ChargeOff ==0
df_x_ChargeOff = df_x[bool_ChargeOff]
df_x_NotChargeOff = df_x[bool_NotChargeOff]
df_x_ChargeOff
```

```python
x,y = [],[]
df_count = df_x_ChargeOff.groupby('ApprovalFY',as_index=False).count()
df_count2 = df_x_NotChargeOff.groupby('ApprovalFY',as_index=False).count()

for i in range(df_count.shape[0]):
    cur_row = df_count.iloc[i]
    df_cur = df_count2[df_count2.ApprovalFY == cur_row.ApprovalFY]
    if df_cur.shape[0]>0:
        nbCharged = df_cur.iloc[0].Name
    else:
        nbCharged = 0
    nbNotCharged = cur_row.Name
    x.append(cur_row.ApprovalFY)
    y.append(nbNotCharged/(nbCharged+nbNotCharged))
    

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
plt.plot(x,y, marker='o', color="blue", alpha=0.3)
plt.title('Percentage of not ssuccessfull loans')
plt.ylabel('Percentage')
plt.xlabel('FY')
```
